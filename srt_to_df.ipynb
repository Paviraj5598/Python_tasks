```python
import pandas as pd
import pysrt
```

```python
def loadSRT(srt_file_path):
        df1 = pd.DataFrame(columns=["Text", "StartTime", "EndTime"])
        try:
            subs = pysrt.open(srt_file_path)
        except Exception as e:
            print(e)
            subs = pysrt.open(srt_file_path, encoding='iso-8859-1')

        for sub in subs:
            tmp = {}
            subText = sub.text
            subText = subText.replace('\n', ' ')
            subText = subText.replace('\t', ' ')
            subText = subText.replace('"', '')
            
            row = {"Text": subText, "StartTime": sub.start.to_time(), "EndTime": sub.end.to_time()}
            df1 = df1.append(row, ignore_index=True)
        return df1
```

```python
final = loadSRT(srt_file_path)
```
